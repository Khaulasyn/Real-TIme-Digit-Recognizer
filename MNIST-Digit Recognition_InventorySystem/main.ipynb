```python
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten, MaxPooling2D
from keras.utils import to_categorical

# Load the MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Preprocess the data
X_train = X_train.reshape(-1, 28, 28, 1) / 255.0
X_test = X_test.reshape(-1, 28, 28, 1) / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)

# Define the model
model = Sequential()
model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D((2, 2)))
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train, y_train, epochs=5, validation_data=(X_test, y_test))

# Save the model
model.save('digit_recognition_model.h5')
```

```python
import cv2
import numpy as np
from skimage.feature import hog
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
import pandas as pd
from collections import deque

# Load the inventory data
file_path = 'inventarisIT.xlsx'
data = pd.read_excel(file_path)

# Extract relevant columns
no_inventaris_col = data.iloc[:, 0].dropna().astype(str).reset_index(drop=True)  # Convert to string and reset index
item_col = data.iloc[:, 1].dropna().reset_index(drop=True)
condition_col = data.iloc[:, 2].dropna().reset_index(drop=True)

# Create a mapping from inventory number to item details
inventory_mapping = {}
for i, no in enumerate(no_inventaris_col):
    inventory_mapping[no] = {'item': item_col[i], 'condition': condition_col[i]}

# Load the MNIST dataset and train the SVM model
def load_and_train_model():
    from sklearn.datasets import fetch_openml

    # Load the dataset
    dataset = fetch_openml('mnist_784')
    features = np.array(dataset.data, dtype='float64')
    labels = np.array(dataset.target, dtype='int')

    # Extract the HOG features for the entire dataset
    list_hog_ft = []
    for feature in features:
        ft = hog(feature.reshape((28, 28)), orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualize=False)
        list_hog_ft.append(ft)
    hog_features = np.array(list_hog_ft, dtype='float64')

    # Fit the StandardScaler with the HOG features
    scaler = StandardScaler()
    hog_features = scaler.fit_transform(hog_features)

    # Fit the LinearSVC with the scaled HOG features
    clf = LinearSVC(max_iter=10000)  # Increased max iterations for better convergence
    clf.fit(hog_features, labels)

    return clf, scaler

clf, scaler = load_and_train_model()

# Function to preprocess the digit image
def preprocess(image):
    if len(image.shape) == 3:  # Check if the image has 3 channels (BGR)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:  # If the image is already in grayscale
        gray = image
    resized = cv2.resize(gray, (28, 28), interpolation=cv2.INTER_AREA)
    reshaped = resized.reshape(1, 28, 28, 1)
    return reshaped / 255.0

# Function to segment the image into individual digits
def segment_image(image):
    if len(image.shape) == 3:  # Check if the image has 3 channels (BGR)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:  # If the image is already in grayscale
        gray = image
    _, thresh = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    digit_images = []
    bounding_boxes = []
    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        # Filter out small or overly large bounding boxes
        if 10 < w < 100 and 10 < h < 100:
            digit = gray[y:y+h, x:x+w]
            digit = cv2.resize(digit, (28, 28), interpolation=cv2.INTER_AREA)
            digit_images.append(digit)
            bounding_boxes.append((x, y, w, h))
    
    # Sort the bounding boxes from left to right
    sorted_boxes = sorted(zip(digit_images, bounding_boxes), key=lambda b: b[1][0])
    digit_images, bounding_boxes = zip(*sorted_boxes) if sorted_boxes else ([], [])
    return digit_images[:13], bounding_boxes[:13]  # Limit to 13 digits

# Rolling window to stabilize predictions
rolling_window = deque(maxlen=10)

# Start the webcam feed
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Segment the frame into individual digit images
    digit_images, bounding_boxes = segment_image(frame)
    
    # Recognize each digit
    recognized_digits = []
    for digit_img in digit_images:
        fd = hog(digit_img, orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualize=False)
        fd = scaler.transform(np.array([fd], 'float64'))
        digit = clf.predict(fd)
        recognized_digits.append(str(digit[0]))
    
    # Combine recognized digits into a number
    recognized_number = ''.join(recognized_digits)

    # Add the recognized number to the rolling window
    rolling_window.append(recognized_number)

    # Determine the most common prediction in the rolling window
    if rolling_window:
        recognized_number = max(set(rolling_window), key=rolling_window.count)

    # Draw bounding boxes and recognized digits on the frame
    for (x, y, w, h), digit in zip(bounding_boxes, recognized_digits):
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, digit, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    # Display the result
    cv2.putText(frame, f'Number: {recognized_number}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Digit Recognition', frame)

    # Check if 'q' or 'esc' is pressed to quit
    key = cv2.waitKey(1) & 0xFF
    if key == ord('q') or key == 27:  # 27 is the Esc key
        break

cap.release()
cv2.destroyAllWindows()

# Debug: Print the recognized number and inventory mapping keys
print(f'Recognized Number: {recognized_number}')
print(f'Inventory Mapping Keys: {list(inventory_mapping.keys())[:5]}')  # Print first 5 keys for reference

# Map the recognized number to the inventory item
if recognized_number in inventory_mapping:
    item = inventory_mapping[recognized_number]
    print(f'Item: {item["item"]}, Condition: {item["condition"]}')
else:
    print('No matching item found.')
```

```python
import cv2
import numpy as np
from skimage.feature import hog
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
import pandas as pd
from collections import deque
from sklearn.datasets import fetch_openml
import gc  # Garbage Collection

# Load the inventory data
file_path = 'inventarisIT.xlsx'
data = pd.read_excel(file_path)

# Extract relevant columns
no_inventaris_col = data.iloc[:, 0].dropna().astype(str).reset_index(drop=True)
item_col = data.iloc[:, 1].dropna().reset_index(drop=True)
condition_col = data.iloc[:, 2].dropna().reset_index(drop=True)

# Create a mapping from inventory number to item details
inventory_mapping = {}
for i, no in enumerate(no_inventaris_col):
    inventory_mapping[no] = {'item': item_col[i], 'condition': condition_col[i]}

# Load and train the SVM model
def load_and_train_model():
    # Load the MNIST dataset
    dataset = fetch_openml('mnist_784', version=1)
    features = np.array(dataset.data, dtype=np.float32)
    labels = np.array(dataset.target, dtype=np.int32)

    # Initialize lists to store HOG features and labels
    list_hog_ft = []
    batch_size = 5000  # Number of samples to process in each batch

    # Process data in chunks to avoid memory issues
    for start in range(0, features.shape[0], batch_size):
        end = min(start + batch_size, features.shape[0])
        batch_features = features[start:end]
        for feature in batch_features:
            ft = hog(feature.reshape((28, 28)), orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualize=False)
            list_hog_ft.append(ft)
            gc.collect()  # Clean up memory

    hog_features = np.array(list_hog_ft, dtype=np.float32)
    
    # Fit the StandardScaler with the HOG features
    scaler = StandardScaler()
    hog_features = scaler.fit_transform(hog_features)

    # Fit the LinearSVC with the scaled HOG features
    clf = LinearSVC(max_iter=10000)
    clf.fit(hog_features, labels)

    return clf, scaler

clf, scaler = load_and_train_model()

# Function to preprocess the digit image
def preprocess(image):
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image
    resized = cv2.resize(gray, (28, 28), interpolation=cv2.INTER_AREA)
    reshaped = resized.reshape(1, 28, 28, 1)
    return reshaped / 255.0

# Function to segment the image into individual digits
def segment_image(image):
    if len(image.shape) == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image
    _, thresh = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    digit_images = []
    bounding_boxes = []
    for cnt in contours:
        if cnt:
            x, y, w, h = cv2.boundingRect(cnt)
            if 10 < w < 100 and 10 < h < 100:
                digit = gray[y:y+h, x:x+w]
                digit = cv2.resize(digit, (28, 28), interpolation=cv2.INTER_AREA)
                digit_images.append(digit)
                bounding_boxes.append((x, y, w, h))
    
    sorted_boxes = sorted(zip(digit_images, bounding_boxes), key=lambda b: b[1][0])
    digit_images, bounding_boxes = zip(*sorted_boxes) if sorted_boxes else ([], [])
    return digit_images[:13], bounding_boxes[:13]  # Limit to 13 digits

# Rolling window to stabilize predictions
rolling_window = deque(maxlen=10)

# Start the webcam feed
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    digit_images, bounding_boxes = segment_image(frame)
    
    recognized_digits = []
    for digit_img in digit_images:
        fd = hog(digit_img, orientations=9, pixels_per_cell=(14, 14), cells_per_block=(1, 1), visualize=False)
        fd = scaler.transform(np.array([fd], dtype=np.float32))
        digit = clf.predict(fd)
        recognized_digits.append(str(digit[0]))
    
    recognized_number = ''.join(recognized_digits)
    rolling_window.append(recognized_number)

    if rolling_window:
        recognized_number = max(set(rolling_window), key=rolling_window.count)

    for (x, y, w, h), digit in zip(bounding_boxes, recognized_digits):
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, digit, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
    
    if bounding_boxes:
        for (x, y, w, h) in bounding_boxes:
            cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    cv2.putText(frame, f'Number: {recognized_number}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Digit Recognition', frame)

    key = cv2.waitKey(1) & 0xFF
    if key == ord('q') or key == 27:
        break

cap.release()
cv2.destroyAllWindows()

print(f'Recognized Number: {recognized_number}')
print(f'Inventory Mapping Keys: {list(inventory_mapping.keys())[:5]}')

if recognized_number in inventory_mapping:
    item = inventory_mapping[recognized_number]
    print(f'Item: {item["item"]}, Condition: {item["condition"]}')
else:
    print('No matching item found.')
```
